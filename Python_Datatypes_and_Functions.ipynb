```python
from typing import AnyStr
# purchase and billing example using return
items = {'rice':56,'chana':30,'wheat':60,'parle_g':10,'milk':35}
bill = []
def purchase(*args):
  for item in args:
    #print(item,items[item])
    bill.append(items[item])
  return bill

def check(total):
  return sum(total)

# calling of purchase
total = purchase('wheat','milk')
print('Item amounts:',total)

# calling of check
ans = check(total)
print('Final bill amoutnt is Rs:',ans)
```

```python
#calculate % and calculate grade
def calculate_per(marks_obtained,total_marks):
  return (marks_obtained/total_marks)*100

#print(calculate_per(450,500))

def calculate_grade(per):
  if per >= 75:
    return 'Distinction'
  elif per >= 65:
    return 'First class'
  elif per>=55:
    return 'Second class'
  elif per >= 45:
      return 'Pass class'
  else:
    return 'Fail'

per=calculate_per(420,500)
result=calculate_grade(82)
print('Total percentage is',per)
print('Your grade is',result)
```
