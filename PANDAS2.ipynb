```python
import numpy as np
import pandas as pd

path = "/content/CO2 Emissions_Canada.csv"

c = pd.read_csv(path)

c
```

```python
c.columns
```

```python
c.select_dtypes(include=float)
```

```python
cat_cols = ['Make', 'Model', 'Vehicle Class', 'Transmission', 'Fuel Type']

for col_name in cat_cols:
  print(col_name,':',c[col_name].nunique())
```

```python
cat_cols = ['Make', 'Model', 'Vehicle Class', 'Transmission', 'Fuel Type']

for col_name in cat_cols:
  print(col_name,':',c[col_name].unique())
```

```python
cat_cols = ['Make', 'Model', 'Vehicle Class', 'Transmission', 'Fuel Type']

for col_name in cat_cols:
  print(col_name,':',c[col_name].value_counts())
```

```python
c['Make'].value_counts()
```

```python
c['Make'].value_counts(sort=False)
```

```python
print("\n\n Value Count")
cat_cols=c.select_dtypes(include='object').columns
for col_name in cat_cols:
  print(col_name,':',c[col_name].value_counts())
  print("="*30)
```

```python
print("\n\n Value Count")
cat_cols=c.select_dtypes(include='object').columns
for col_name in cat_cols:
  print(col_name,':',c[col_name].value_counts().head(3))
  print("="*30)
```
